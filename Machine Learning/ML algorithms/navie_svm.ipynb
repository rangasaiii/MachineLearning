```python
import pandas as pd
```

```python
df=pd.read_csv("/content/adult (2).csv")
```

```python
df.head()
```

```python
#Name of the columns
df.columns = ['Age', 'Workclass', 'fnlgt', 'Education', 'Education_num', 'Marital_Status', 'Occupation'
               ,'Relationship', 'Race', 'Sex', 'Gain', 'Loss', 'Hoursperweek', 'Country', 'Label']
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['Label']=le.fit_transform(df['Label'])
```

```python
df.head()
```

```python
num=[var for var in df.columns if (df[var].dtype=='int64') & (var!='Label')]
cat=[ var for var in df.columns if(df[var].dtype=='object')]
```

```python
num
```

```python
cat
```

```python
from scipy.stats import zscore
```

```python
normalized_data=zscore(df[num],axis=1)
normalized_data
```

```python

import numpy as np
np.mean(normalized_data)
```

```python
np.std(normalized_data)
```

```python
df=pd.concat([normalized_data,df[cat],df['Label']],axis=1)
```

```python
df.head()
```

```python
# cardinality
for var in cat:
  print(var+ ' contains '+str(df[var].nunique())+' labels')
  print()
```

```python
df=pd.get_dummies(df,columns=cat)
df
```

```python
x=df.drop(columns='Label',axis=1)
y=df['Label']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=100)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
y_pred=gnb.predict(x_test)
```
